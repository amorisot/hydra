```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

PATHS  = 'stats_pickled/stat_summaries_vgg.pkl'

with open(PATHS, 'rb') as pickled:
    a = pickle.load(pickled)
```

```python
print(list(a.keys()))
```

```python
xs = np.arange(200)+1
```

```python
acc_hydra = np.sum(a['hydra accuracy'], axis=0)/5
acc_classifier = np.sum(a['classifier accuracy'], axis=0)/5
```

```python
plt.plot(xs, acc_hydra, c='r')
plt.plot(xs, acc_classifier, c='g')
plt.show()
```

```python
hydra_test = np.sum(a['hydra test loss'], axis=0)/5
hydra_train = np.sum(a['hydra train loss'], axis=0)/5
classifier_test = np.sum(a['classifier test loss'], axis=0)/5
classifier_train = np.sum(a['classifier train loss'], axis=0)/5
```

```python
plt.plot(xs, hydra_test, c='r')
plt.plot(xs, hydra_train, c='pink')
plt.plot(xs, classifier_test, c='g')
plt.plot(xs, classifier_train, c='lime')
```
